```python
# 排除负值
df1 = df[df.STDDEV1>0]

std_STDDEV1 = df1['STDDEV1'].std()
mean_STDDEV1 = df1['STDDEV1'].mean() 
kf_STDDEV1 = np.sqrt(std_STDDEV1)

STDDEV1_highline_1 = round(mean_STDDEV1 + kf_STDDEV1,2)
STDDEV1_lowline_1 = round(mean_STDDEV1 - kf_STDDEV1,2)

STDDEV1_1 = [STDDEV1_lowline_1,STDDEV1_highline_1]

STDDEV1_highline_2 = round(mean_STDDEV1 + std_STDDEV1,2)
STDDEV1_lowline_2 = round(mean_STDDEV1 - std_STDDEV1,2)

STDDEV1_2 = [STDDEV1_lowline_2,STDDEV1_highline_2]

print( 'STDDEV1 第一阶梯: ' + str(STDDEV1_1))
print( 'STDDEV1 第二阶梯: ' + str(STDDEV1_2))
```
